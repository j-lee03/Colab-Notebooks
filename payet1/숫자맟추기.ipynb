```python
import random

def number_guessing_game():
    """
    1ы╢АэД░ 100ъ╣МьзАьЭШ ьИльЮРые╝ ызЮь╢ФыКФ ъ▓МьЮДьЮЕыЛИыЛд.
    ь┤Э 10ы▓ИьЭШ ъ╕░эЪМъ░А ьг╝ьЦ┤ьзСыЛИыЛд.
    """
    secret_number = random.randint(1, 100)
    max_attempts = 10

    print("--- ьИльЮР ызЮь╢Фъ╕░ ъ▓МьЮДьЧР ьШдьЛа ъ▓ГьЭД эЩШьШБэХйыЛИыЛд! ---")
    print("ьаАыКФ 1ы╢АэД░ 100 ьВмьЭ┤ьЭШ ы╣Ды░А ьИльЮРые╝ эХШыВШ ьГЭъ░БэЦИьЦ┤ьЪФ. ызЮь╢░ы│┤ьД╕ьЪФ!")
    print(f"ъ╕░эЪМыКФ ь┤Э {max_attempts}ы▓И ьг╝ьЦ┤ьзСыЛИыЛд.")

    for attempt in range(1, max_attempts + 1):
        print(f"\n[{attempt}ы▓Иьз╕ ьЛЬыПД]")
        try:
            guess = int(input("ьИльЮРые╝ ьЮЕыаеэХШьД╕ьЪФ: "))

            if guess < 1 or guess > 100:
                print("1ы╢АэД░ 100 ьВмьЭ┤ьЭШ ьИльЮРые╝ ьЮЕыаеэХ┤ ьг╝ьД╕ьЪФ.")
                continue

            if guess == secret_number:
                print(f"ЁЯОЙЁЯОЙЁЯОЙ ы╣аы░Фы░д! ьаХыЛ╡ьЮЕыЛИыЛд! ЁЯОЙЁЯОЙЁЯОЙ")
                print(f"ь╢ХэХШэХйыЛИыЛд! {attempt}ы▓И ызМьЧР ызЮь╢ФьЕиьЦ┤ьЪФ!")
                return
            elif guess < secret_number:
                print("ыНФ эБ░ ьИльЮРьШИьЪФ.")
            else:
                print("ыНФ ьЮСьЭА ьИльЮРьШИьЪФ.")
        except ValueError:
            print("ьШмы░Фые╕ ьИльЮР эШХьЛЭьЭ┤ ьХДыЛЩыЛИыЛд. ыЛдьЛЬ ьЛЬыПДэХ┤ ьг╝ьД╕ьЪФ.")

    print("\n--- ъ▓МьЮД ьвЕыгМ ---")
    print(f"ьЧРэЬ┤ ьЭ┤ъ▓ГыПД ыкиые┤ыГР уЕЛуЕЛуЕЛ")
    print(f"ьаХыЛ╡ьЭА {secret_number}ьЭ┤ьЧИьЦ┤ьЪФ.")

# ъ▓МьЮД ьЛЬьЮС
if __name__ == "__main__":
    number_guessing_game()
```

```python
import random

def generate_lotto():
    """ыбЬыШР ы▓ИэШ╕ 6ъ░Ьые╝ ьГЭьД▒эХШыКФ эХиьИШ"""
    # random.sampleыбЬ ьдСы│╡ ьЧЖьЭ┤ 6ъ░Ь ьДаэГЭ
    numbers = random.sample(range(1, 46), 6)  # 1~45ьЧРьДЬ 6ъ░Ь ьДаэГЭ
    numbers.sort()  # ьШдыжДь░иьИЬ ьаХыам
    return numbers

print("ЁЯО░ ыбЬыШР ы▓ИэШ╕ ьГЭьД▒ъ╕░")
print("=" * 30)

# ьВмьЪйьЮРъ░А ьЫРэХШыКФ ызМэБ╝ ьГЭьД▒
while True:
    try:
        count = int(input("ыкЗ ъ░ЬьЭШ ы▓ИэШ╕ые╝ ьГЭьД▒эХаъ╣МьЪФ? (0ьЭА ьвЕыгМ): "))

        if count == 0:
            print("эФДыбЬъ╖╕ыЮиьЭД ьвЕыгМэХйыЛИыЛд. ЁЯСЛ")
            break
        elif count < 0:
            print("ьЦСьИШые╝ ьЮЕыаеэХ┤ьг╝ьД╕ьЪФ!")
            continue

        print(f"\nЁЯОп {count}ъ░ЬьЭШ ыбЬыШР ы▓ИэШ╕:")
        print("-" * 25)

        for i in range(count):
            lotto_numbers = generate_lotto()
            print(f"{i+1}ы▓Иьз╕: {lotto_numbers}")
            # ьШИьГБ ъ▓░ъ│╝: [3, 15, 22, 31, 38, 44] (ызды▓И ыЛдыжД)

        print(f"\nь┤Э {count}ъ░ЬьЭШ ы▓ИэШ╕ъ░А ьГЭьД▒ыРШьЧИьК╡ыЛИыЛд!")
        print("эЦЙьЪ┤ьЭД ы╣ХыЛИыЛд! ЁЯНА\n")
        print("эФДыбЬъ╖╕ыЮиьЭД ьвЕыгМэХйыЛИыЛд. ЁЯСЛ")
        break

    except ValueError:
        print("ьИльЮРые╝ ьЮЕыаеэХ┤ьг╝ьД╕ьЪФ!")
```

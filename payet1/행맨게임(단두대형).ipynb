```python
import random

def hangman_with_hint():
    hangman_pics = [
        """
           +---+
           |   |
               |
               |
               |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
               |
               |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
           |   |
               |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
          /|   |
               |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
          /|\\  |
               |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
          /|\\  |
          /    |
               |
        =========
        """,
        """
           +---+
           |   |
           O   |
          /|\\  |
          / \\  |
               |
        =========
        """
    ]
    words = ["아버지", "어머니", "오즈스쿨은", "너무나", "좋은코딩", "부트캠프", "인거갔아요", "오즈스쿨을", "추천합니다"]
    chosen_word = random.choice(words)
    word_length = len(chosen_word)
    lives = 10
    guessed_letters = []
    display = ["_"] * word_length
    game_over = False
    hints_left = 5
    print("행맨 게임에 오신 것을 환영합니다!")
    print("힌트를 얻고 싶으면 '힌트'를 입력하세요. (기회는 단 한 번!)")
    while not game_over:
        print(hangman_pics[10 - lives])
        print(f"현재 단어: {' '.join(display)}")
        print(f"남은 목숨: {lives}")
        print(f"추측한 글자들: {', '.join(guessed_letters)}")
        guess = input("한 글자를 추측하거나 '힌트'를 입력하세요: ").lower()
        if guess == "힌트":
            if hints_left > 0:
                unrevealed_letters = [letter for letter in chosen_word if letter not in guessed_letters]
                if unrevealed_letters:
                    hint_letter = random.choice(unrevealed_letters)
                    print(f"💡 힌트: 글자 '{hint_letter}'가 단어에 포함되어 있습니다!")
                    for i in range(word_length):
                        if chosen_word[i] == hint_letter:
                            display[i] = hint_letter

                    if hint_letter not in guessed_letters:
                         guessed_letters.append(hint_letter)

                    hints_left -= 1
                else:
                    print("모든 글자를 거의 다 맞추셨네요! 힌트가 필요 없습니다.")
            else:
                print("힌트 기회를 이미 모두 사용했습니다.")
            continue
        if guess in guessed_letters:
            print(f"이미 '{guess}'는 추측한 글자입니다. 다른 글자를 시도해보세요.")
            continue
        guessed_letters.append(guess)
        if guess in chosen_word:
            for i in range(word_length):
                if chosen_word[i] == guess:
                    display[i] = guess
        else:
            print(f"'{guess}'는 단어에 포함되어 있지 않습니다.")
            lives -= 1
        if "_" not in display:
            game_over = True
            print("\n🎉 축하합니다! 단어를 모두 맞추셨습니다!")
            print(f"정답은 '{chosen_word}'였습니다.")
        if lives == 0:
            game_over = True
            print(hangman_pics[6])
            print("\n이런, 목숨을 모두 사용했습니다. 게임 오버! 💀")
            print(f"정답은 '{chosen_word}'였습니다.")

if __name__ == "__main__":
    hangman_with_hint()
```

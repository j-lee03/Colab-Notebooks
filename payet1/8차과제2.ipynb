2. 거래 정보를 문자열로 반환하는 __str__ 메서드를 구현합니다.
3. 거래 정보를 튜플로 반환하는 to_tuple 메서드를 구현합니다.

<b>변수 컨벤션:</b>

- transaction_type: 거래 유형을 나타내는 문자열 (예: "입금", "출금")
- amount: 거래 금액을 나타내는 정수
- balance: 거래 후 잔고를 나타내는 정수

##### 문제 1.1: 생성자 구현

- transaction_type, amount, balance 속성을 초기화하는 생성자를 구현하세요.
- 타입 힌팅을 사용하여 생성자의 매개변수 타입을 지정하세요.
    - 함수 시그니처: __init__(self, transaction_type: str, amount: int, balance: int) -> None

```python
class Transaction:
    def __init__(self, transaction_type: str, amount: int, balance: int) -> None:
        self.transaction_type = transaction_type
        self.amount = amount
        self.balance = balance
    def to_tuple(self) -> tuple:
        return (self.transaction_type, self.amount, self.balance)
def __str__(self) -> str:
        return f"유형: {self.transaction_type}, 금액: {self.amount}, 거래 후 잔고: {self.balance}"
def to_tuple(self) -> tuple:
        return (self.transaction_type, self.amount, self.balance)
```

##### 문제 1.2: 문자열 반환 메서드 구현

- 타입 힌팅을 사용하여 메서드의 반환 타입을 지정하세요.
    - 함수 시그니처: __str__(self) -> str

##### 문제 1.3: 튜플 반환 메서드 구현

- 거래 정보를 튜플로 반환하는 to_tuple 메서드를 구현하세요.
- 타입 힌팅을 사용하여 메서드의 반환 타입을 지정하세요.
    - 함수 시그니처: to_tuple(self) -> tuple

```python
class Transaction:
    def __init__(self, transaction_type: str, amount: int, balance: int):
        self.transaction_type = transaction_type
        self.amount = amount
        self.balance = balance
    def __str__(self) -> str:
        return f"유형: {self.transaction_type}, 금액: {self.amount}, 거래 후 잔고: {self.balance}"
class Account:
    bank_name = "멋쟁이 은행"
    def __init__(self) -> None:
        self.__balance = 0
        self.transactions = []
    def deposit(self, amount: int) -> None:
        if amount > 0:
            self.__balance += amount
            self.transactions.append(Transaction("입금", amount, self.__balance))
            print(f"{amount}원 입금 완료. 현재 잔액: {self.__balance}원")
        else:
            print("오류: 입금액은 0보다 커야 합니다.")
    def withdraw(self, amount: int) -> None:
        if amount <= 0:
            print("오류: 출금액은 0보다 커야 합니다.")
        elif self.__balance < amount:
            print(f"오류: 잔고 부족. 현재 잔액: {self.__balance}원")
        else:
            self.__balance -= amount
            self.transactions.append(Transaction("출금", amount, self.__balance))
            print(f"{amount}원 출금 완료. 현재 잔액: {self.__balance}원")
    def get_balance(self) -> int:
        return self.__balance
    def get_transactions(self) -> list:
        return self.transactions
    @classmethod
    def get_bank_name(cls) -> str:
        return cls.bank_name
    @classmethod
    def set_bank_name(cls, name: str) -> None:
        cls.bank_name = name
```

```python
class Transaction:
    def __init__(self, tt, a, b): self.tt, self.a, self.b = tt, a, b
    def __str__(self): return f"유형:{self.tt}, 금액:{self.a}, 잔고:{self.b}"
class Account:
    def __init__(self): self.b, self.t = 0, []
    def deposit(self, a): self.b += a; self.t.append(Transaction("입금", a, self.b)); print(f"입금:{a}")
    def withdraw(self, a):
        if self.b >= a: self.b -= a; self.t.append(Transaction("출금", a, self.b)); print(f"출금:{a}")
        else: print("잔고부족")
    def get_balance(self): return self.b
    def get_transactions(self): return self.t
class User:
    def __init__(self, u): self.username, self.account = u, Account()

class UserNotFoundError(Exception): pass

class BankingService:
    def __init__(self) -> None:
        self.users = []
    def add_user(self, username: str) -> None:
        user = User(username)
        self.users.append(user)
        print(f"'{username}'님, 환영합니다! 계좌가 생성되었습니다.")
    def find_user(self, username: str) -> User:
        for user in self.users:
            if user.username == username:
                return user
        raise UserNotFoundError(f"사용자를 찾을 수 없습니다: {username}")
    def user_menu(self, username: str) -> None:
        try:
            user = self.find_user(username)
            while True:
                print(f"\n--- [{user.username}]님 메뉴 ---")
                print("1. 입금 | 2. 출금 | 3. 잔고 확인 | 4. 거래 내역 | 5. 로그아웃")
                choice = input("선택: ")
                if choice == '1':
                    try:
                        amount = int(input("입금할 금액: "))
                        user.account.deposit(amount)
                    except ValueError:
                        print("오류: 숫자를 입력해주세요.")
                elif choice == '2':
                    try:
                        amount = int(input("출금할 금액: "))
                        user.account.withdraw(amount)
                    except ValueError:
                        print("오류: 숫자를 입력해주세요.")
                elif choice == '3':
                    print(f"현재 잔고: {user.account.get_balance()}원")
                elif choice == '4':
                    transactions = user.account.get_transactions()
                    if not transactions:
                        print("거래 내역이 없습니다.")
                    else:
                        for t in transactions:
                            print(t)
                elif choice == '5':
                    print("로그아웃합니다.")
                    break
                else:
                    print("잘못된 선택입니다.")
        except UserNotFoundError as e:
            print(f"오류: {e}")
```

##### 문제 5.1: 데코레이터 구현

- validate_transaction 데코레이터를 작성하여 금액이 0보다 큰지 확인합니다.
- 타입 힌팅을 사용하여 함수의 매개변수와 반환 타입을 지정하세요.
    - 함수 시그니처: validate_transaction(func: Callable) -> Callable

```python
from typing import Callable
class InsufficientFundsError(Exception):
    def __init__(self, balance: int, amount: int) -> None:
        super().__init__(f"잔고 부족. 현재 잔액: {balance}, 출금 요청액: {amount}")
class NegativeAmountError(Exception):
    def __init__(self) -> None:
        super().__init__("금액은 0보다 커야 합니다.")
class UserNotFoundError(Exception):
    def __init__(self, username: str) -> None:
        super().__init__(f"사용자를 찾을 수 없습니다: {username}")
def validate_transaction(func: Callable) -> Callable:
    def wrapper(self, amount: int):
        if amount <= 0:
            raise NegativeAmountError()
        return func(self, amount)
    return wrapper
class TestAccount:
    balance = 100
    @validate_transaction
    def deposit(self, amount: int):
        print(f"{amount}원 입금을 처리합니다.")
    @validate_transaction
    def withdraw(self, amount: int):
        if self.balance < amount:
            raise InsufficientFundsError(self.balance, amount)
        print(f"{amount}원 출금을 처리합니다.")
print("--- 데코레이터 및 예외 처리 테스트 ---")
test_acc = TestAccount()
try:
    test_acc.deposit(50)
    test_acc.withdraw(200)
except InsufficientFundsError as e:
    print(f"오류 발생: {e}")
try:
    test_acc.deposit(-10)
except NegativeAmountError as e:
    print(f"오류 발생: {e}")
```

##### 문제 6.1: 메인 함수 구현

- BankingService 인스턴스를 생성하세요.
- 사용자로부터 입력을 받아 사용자 추가 및 찾기 기능을 구현하세요.
- 사용자 메뉴를 통해 입금, 출금, 잔고 확인, 거래 내역 기능을 실행할 수 있도록 구현하세요.
- 타입 힌팅을 사용하여 함수의 매개변수와 반환 타입을 지정하세요.
    - 함수 시그니처: main() -> None

```python
import time
import sys
from typing import Callable
def processing_animation(duration=1):
    animation = "|/-\\"
    start_time = time.time()
    while time.time() < start_time + duration:
        for char in animation:
            sys.stdout.write(f"\r처리 중... {char}")
            sys.stdout.flush()
            time.sleep(0.1)
    sys.stdout.write("\r완료!          \n")
class InsufficientFundsError(Exception):
    def __init__(self, balance: int, amount: int): super().__init__(f"잔고 부족. 잔액:{balance}, 요청액:{amount}")
class NegativeAmountError(Exception):
    def __init__(self): super().__init__("금액은 0보다 커야 합니다.")
class UserNotFoundError(Exception):
    def __init__(self, username: str): super().__init__(f"사용자 없음: {username}")
def validate_transaction(func: Callable) -> Callable:
    def wrapper(self, amount: int):
        if amount <= 0: raise NegativeAmountError()
        return func(self, amount)
    return wrapper
class Transaction:
    def __init__(self, tt: str, a: int, b: int): self.tt, self.a, self.b = tt, a, b
    def __str__(self): return f"유형: {self.tt}, 금액: {self.a}, 잔고: {self.b}"
class Account:
    bank_name = "멋쟁이 은행"
    def __init__(self): self.__balance, self.transactions = 0, []
    @validate_transaction
    def deposit(self, a: int):
        print("입금을 진행합니다...")
        processing_animation()
        self.__balance += a
        self.transactions.append(Transaction("입금", a, self.__balance))
        print(f"✅ {a}원 입금 완료. 현재 잔액: {self.__balance}원")
    @validate_transaction
    def withdraw(self, a: int):
        print("출금을 진행합니다...")
        if self.__balance < a: 
            raise InsufficientFundsError(self.__balance, a)
        processing_animation()
        self.__balance -= a
        self.transactions.append(Transaction("출금", a, self.__balance))
        print(f"✅ {a}원 출금 완료. 현재 잔액: {self.__balance}원")
    def get_balance(self): return self.__balance
    def get_transactions(self): return self.transactions
    @classmethod
    def get_bank_name(cls): return cls.bank_name
class User:
    def __init__(self, u: str): self.username, self.account = u, Account()
class BankingService:
    def __init__(self): self.users = []
    def add_user(self, u: str):
        print(f"'{u}'님 계좌를 생성합니다...")
        processing_animation()
        self.users.append(User(u))
        print(f"🎉 '{u}'님, 계좌 생성이 완료되었습니다.")
    def find_user(self, u: str):
        for user in self.users:
            if user.username == u: return user
        raise UserNotFoundError(u)
    def user_menu(self, u: str):
        try:
            print(f"'{u}'님 정보를 찾는 중...")
            user = self.find_user(u)
            processing_animation(0.5)
            while True:
                print(f"\n--- [{user.username}]님 메뉴 ---")
                print("1.입금 | 2.출금 | 3.잔고 | 4.거래내역 | 5.로그아웃")
                c = input("선택: ")
                try:
                    if c == '1': user.account.deposit(int(input("입금액: ")))
                    elif c == '2': user.account.withdraw(int(input("출금액: ")))
                    elif c == '3': print(f"현재 잔고: {user.account.get_balance()}원")
                    elif c == '4':
                        txs = user.account.get_transactions()
                        print("\n--- 거래 내역 ---")
                        if not txs: print("거래 내역이 없습니다.")
                        else: [print(t) for t in txs]
                        print("-----------------")
                    elif c == '5': print("로그아웃합니다."); break
                    else: print("잘못된 선택입니다.")
                except (ValueError, NegativeAmountError, InsufficientFundsError) as e: 
                    print(f"오류: {e}")
        except UserNotFoundError as e: 
            print(f"오류: {e}")
def main() -> None:
    service = BankingService()
    print(f"** {Account.get_bank_name()} 온라인 뱅킹 시스템 **")
    while True:
        print("\n--- 메인 메뉴 ---")
        print("1. 사용자 추가 | 2. 사용자 로그인 | 3. 종료")
        choice = input("선택: ")
        if choice == '1': service.add_user(input("사용자 이름: "))
        elif choice == '2': service.user_menu(input("로그인 이름: "))
        elif choice == '3': print("시스템 종료."); break
        else: print("잘못된 선택.")
if __name__ == "__main__":
    main()
```

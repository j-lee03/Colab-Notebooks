```python
from typing import Callable
class InsufficientFundsError(Exception):
    def __init__(self, balance: int, amount: int) -> None:
        super().__init__(f"ì”ê³  ë¶€ì¡±. í˜„ì¬ ì”ì•¡: {balance}, ì¶œê¸ˆ ìš”ì²­ì•¡: {amount}")
class NegativeAmountError(Exception):
    def __init__(self) -> None:
        super().__init__("ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤.")
class UserNotFoundError(Exception):
    def __init__(self, username: str) -> None:
        super().__init__(f"ì‚¬ìš©ìë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {username}")
def validate_transaction(func: Callable) -> Callable:
    def wrapper(self, amount: int):
        if amount <= 0:
            raise NegativeAmountError()
        return func(self, amount)
    return wrapper
class TestAccount:
    balance = 100
    @validate_transaction
    def deposit(self, amount: int):
        print(f"{amount}ì› ì…ê¸ˆì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.")
    @validate_transaction
    def withdraw(self, amount: int):
        if self.balance < amount:
            raise InsufficientFundsError(self.balance, amount)
        print(f"{amount}ì› ì¶œê¸ˆì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.")
print("--- ë°ì½”ë ˆì´í„° ë° ì˜ˆì™¸ ì²˜ë¦¬ í…ŒìŠ¤íŠ¸ ---")
test_acc = TestAccount()
try:
    test_acc.deposit(50)
    test_acc.withdraw(200)
except InsufficientFundsError as e:
    print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")
try:
    test_acc.deposit(-10)
except NegativeAmountError as e:
    print(f"ì˜¤ë¥˜ ë°œìƒ: {e}")
```

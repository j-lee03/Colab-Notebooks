__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
print("===== 붕어빵 주문 =====")
for name, stock in inventory.items():
    print(f"- {name}: {stock}개")
print("======================")

order_item = input("주문할 붕어빵 종류를 입력하세요 (예: 팥붕어빵): ")
order_quantity = int(input(f"{order_item}을(를) 몇 개 주문하시겠어요?: "))
if order_item in inventory and inventory[order_item] >= order_quantity:
    inventory[order_item] -= order_quantity
    print(f"\n✅ 판매 완료! {order_item} {order_quantity}개를 판매했습니다.")
    print("감사합니다!")
elif order_item in inventory and inventory[order_item] < order_quantity:
    print(f"\n🚨 죄송합니다. {order_item}의 재고가 부족합니다. (남은 재고: {inventory[order_item]}개)")
else:
    print(f"\n🚨 죄송합니다. '{order_item}'은(는) 판매하지 않는 메뉴입니다.")
print("\n===== 현재 남은 재고 =====")
for name, stock in inventory.items():
    print(f"- {name}: {stock}개")
print("==========================")
```

__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
while True:
    print("\n===== 붕어빵 주문 =====")
    for name, stock in inventory.items():
        print(f"- {name}: {stock}개")
    print("======================")

    order_item = input("주문할 붕어빵 종류를 입력하세요 (종료하시려면 '종료' 입력): ")
    if order_item == "종료":
        print("\n프로그램을 종료합니다. 이용해주셔서 감사합니다.")
        break
    if order_item not in inventory:
        print(f"\n🚨 죄송합니다. '{order_item}'은(는) 판매하지 않는 메뉴입니다.")
        continue 
    try:
        order_quantity = int(input(f"{order_item}을(를) 몇 개 주문하시겠어요?: "))
        if order_quantity <= 0:
            print("\n🚨 1개 이상 주문해주세요.")
            continue
    except ValueError:
        print("\n🚨 잘못된 입력입니다. 수량은 숫자로 입력해주세요.")
        continue
    if inventory[order_item] >= order_quantity:
        inventory[order_item] -= order_quantity
        print(f"\n✅ 판매 완료! {order_item} {order_quantity}개를 판매했습니다.")
        print("감사합니다!")
    else:
        print(f"\n🚨 죄송합니다. {order_item}의 재고가 부족합니다. (남은 재고: {inventory[order_item]}개)")
```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
while True:
    print("\n*** 붕어빵 자동 판매 시스템 ***")
    mode = input("모드를 선택하세요 (1: 손님, 2: 관리자, 3: 종료): ")
    if mode == '1':
        while True:
            print("\n===== 붕어빵 주문 =====")
            for name, stock in inventory.items():
                print(f"- {name}: {stock}개")
            print("======================")
            order_item = input("주문할 붕어빵 종류를 입력하세요 (이전으로 돌아가려면 '종료' 입력): ")
            if order_item == "종료":
                break
            if order_item not in inventory:
                print(f"\n🚨 죄송합니다. '{order_item}'은(는) 판매하지 않는 메뉴입니다.")
                continue
            try:
                order_quantity = int(input(f"{order_item}을(를) 몇 개 주문하시겠어요?: "))
                if order_quantity <= 0:
                    print("\n🚨 1개 이상 주문해주세요.")
                    continue
            except ValueError:
                print("\n🚨 잘못된 입력입니다. 수량은 숫자로 입력해주세요.")
                continue
            if inventory[order_item] >= order_quantity:
                inventory[order_item] -= order_quantity
                print(f"\n✅ 판매 완료! {order_item} {order_quantity}개를 판매했습니다.")
                print("감사합니다!")
            else:
                print(f"\n🚨 죄송합니다. {order_item}의 재고가 부족합니다. (남은 재고: {inventory[order_item]}개)")
    elif mode == '2':
        while True:
            print("\n===== 관리자 모드: 재고 추가 =====")
            for name, stock in inventory.items():
                print(f"- {name}: {stock}개")
            print("===============================")
            item_to_add = input("재고를 추가할 붕어빵 종류를 입력하세요 (이전으로 돌아가려면 '종료' 입력): ")
            if item_to_add == "종료":
                break
            if item_to_add not in inventory:
                print(f"\n🚨 '{item_to_add}'은(는) 등록된 메뉴가 아닙니다.")
                continue
            try:
                quantity_to_add = int(input(f"{item_to_add}의 재고를 몇 개 추가하시겠어요?: "))
                if quantity_to_add <= 0:
                    print("\n🚨 1개 이상 추가해주세요.")
                    continue
            except ValueError:
                print("\n🚨 잘못된 입력입니다. 수량은 숫자로 입력해주세요.")
                continue
            inventory[item_to_add] += quantity_to_add
            print(f"\n✅ 재고 추가 완료! {item_to_add}의 현재 재고는 {inventory[item_to_add]}개입니다.")
    elif mode == '3':
        print("\n프로그램을 종료합니다. 이용해주셔서 감사합니다.")
        break
    else:
        print("\n🚨 잘못된 입력입니다. 1, 2, 3 중에서 선택해주세요.")
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
inventory = {
    "팥붕어빵": 10,
    "슈크림붕어빵": 8,
    "초코붕어빵": 5
}
prices = {
    "팥붕어빵": 1000,
    "슈크림붕어빵": 1200,
    "초코붕어빵": 1500
}
sales = {
    "팥붕어빵": 0,
    "슈크림붕어빵": 0,
    "초코붕어빵": 0
}
while True:
    print("\n*** 붕어빵 자동 판매 시스템 ***")
    mode = input("모드를 선택하세요 (1: 손님, 2: 관리자, 3: 종료): ")
    if mode == '1':
        while True:
            print("\n===== 붕어빵 주문 =====")
            for name, stock in inventory.items():
                print(f"- {name}: {stock}개 (가격: {prices[name]}원)")
            print("======================")
            order_item = input("주문할 붕어빵 종류를 입력하세요 (이전으로 돌아가려면 '종료' 입력): ")
            if order_item == "종료":
                break
            if order_item not in inventory:
                print(f"\n🚨 죄송합니다. '{order_item}'은(는) 판매하지 않는 메뉴입니다.")
                continue
            try:
                order_quantity = int(input(f"{order_item}을(를) 몇 개 주문하시겠어요?: "))
                if order_quantity <= 0:
                    print("\n🚨 1개 이상 주문해주세요.")
                    continue
            except ValueError:
                print("\n🚨 잘못된 입력입니다. 수량은 숫자로 입력해주세요.")
                continue
            if inventory[order_item] >= order_quantity:
                inventory[order_item] -= order_quantity
                sales[order_item] += order_quantity
                print(f"\n✅ 판매 완료! {order_item} {order_quantity}개를 판매했습니다.")
                print("감사합니다!")
            else:
                print(f"\n🚨 죄송합니다. {order_item}의 재고가 부족합니다. (남은 재고: {inventory[order_item]}개)")
    elif mode == '2':
        while True:
            print("\n===== 관리자 모드: 재고 추가 =====")
            for name, stock in inventory.items():
                print(f"- {name}: {stock}개")
            print("===============================")
            item_to_add = input("재고를 추가할 붕어빵 종류를 입력하세요 (이전으로 돌아가려면 '종료' 입력): ")
            if item_to_add == "종료":
                break
            if item_to_add not in inventory:
                print(f"\n🚨 '{item_to_add}'은(는) 등록된 메뉴가 아닙니다.")
                continue
            try:
                quantity_to_add = int(input(f"{item_to_add}의 재고를 몇 개 추가하시겠어요?: "))
                if quantity_to_add <= 0:
                    print("\n🚨 1개 이상 추가해주세요.")
                    continue
            except ValueError:
                print("\n🚨 잘못된 입력입니다. 수량은 숫자로 입력해주세요.")
                continue
            inventory[item_to_add] += quantity_to_add
            print(f"\n✅ 재고 추가 완료! {item_to_add}의 현재 재고는 {inventory[item_to_add]}개입니다.")
    elif mode == '3':
        print("\n프로그램을 종료합니다. 이용해주셔서 감사합니다.")
        print("\n===== 최종 매출 보고 =====")
        total_revenue = 0
        for name, count in sales.items():
            revenue = prices[name] * count
            total_revenue += revenue
            print(f"- {name}: {count}개 판매, 매출: {revenue}원")
        print("--------------------------")
        print(f"✨ 총 매출: {total_revenue}원")
        print("==========================")
        break
    else:
        print("\n🚨 잘못된 입력입니다. 1, 2, 3 중에서 선택해주세요.")
```

### íŒŒì¼ ì²˜ë¦¬ ì˜ˆì œ

````python
# í•™ìƒ ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ
def save_scores(filename, students_data):
    """í•™ìƒ ì„±ì ì„ íŒŒì¼ì— ì €ì¥"""
    with open(filename, 'w', encoding='utf-8') as file:
        file.write("ì´ë¦„,ìˆ˜í•™,ì˜ì–´,ê³¼í•™\n")  # í—¤ë”
        for name, scores in students_data.items():
            line = f"{name},{scores['math']},{scores['english']},{scores['science']}\n"
            file.write(line)

def load_scores(filename):
    """íŒŒì¼ì—ì„œ í•™ìƒ ì„±ì  ë¶ˆëŸ¬ì˜¤ê¸°"""
    students = {}
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            for line in lines[1:]:  # í—¤ë” ì œì™¸
                data = line.strip().split(',')
                name = data[0]
                students[name] = {
                    'math': int(data[1]),
                    'english': int(data[2]),
                    'science': int(data[3])
                }
    except FileNotFoundError:
        print("íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    return students
````

````python
# í•™ìƒ ë°ì´í„°
students_data = {
    'ê¹€ì² ìˆ˜': {'math': 85, 'english': 92, 'science': 78},
    'ì´ì˜í¬': {'math': 90, 'english': 85, 'science': 95}
}
````

````python
# íŒŒì¼ ì“°ê¸°
save_scores('scores.txt', students_data)
````

````python
# íŒŒì¼ ì½ê¸°
loaded_data = load_scores('scores.txt')
loaded_data
````

### íŒŒì¼ ê´€ë ¨ ì˜ˆì™¸ì²˜ë¦¬

````python
def safe_file_read(filename):
    """ì•ˆì „í•œ íŒŒì¼ ì½ê¸° í•¨ìˆ˜"""
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except FileNotFoundError:
        print(f"'{filename}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None
    except PermissionError:
        print(f"'{filename}' íŒŒì¼ì— ì ‘ê·¼í•  ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤.")
        return None
    except UnicodeDecodeError:
        print(f"'{filename}' íŒŒì¼ì˜ ì¸ì½”ë”©ì„ ì½ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None
    except Exception as e:
        print(f"ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
        return None

def safe_file_write(filename, content):
    """ì•ˆì „í•œ íŒŒì¼ ì“°ê¸° í•¨ìˆ˜"""
    try:
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(content)
        print(f"'{filename}' íŒŒì¼ì— ì„±ê³µì ìœ¼ë¡œ ì €ì¥í–ˆìŠµë‹ˆë‹¤.")
        return True
    except PermissionError:
        print(f"'{filename}' íŒŒì¼ì— ì“¸ ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤.")
        return False
    except Exception as e:
        print(f"íŒŒì¼ ì €ì¥ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return False
````

````python
# ì‚¬ìš© ì˜ˆì‹œ
content = safe_file_read('diary.txt')
if content:
    print("íŒŒì¼ ë‚´ìš©:", content)

success = safe_file_write('output.txt', "Hello, World!")
if success:
    print("ì €ì¥ ì™„ë£Œ!")
````

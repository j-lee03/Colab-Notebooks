__[문제4]__

3일 뒤 차에 남은 연료는 몇 리터인가요?

```python
a= 9
b = 3
c = 2
d = b * c
e = a - d
print(f"초기 연료량: {a}리터")
print(f"총 운행일: {b}일")
print(f"하루 연료 소비량: {c}리터")
print("-" * 20)
print(f"3일 뒤 남은 연료는 {e}리터입니다.")
```
